Importing Dependecies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data Collection and Processing

```python
car_dataset = pd.read_csv('Car Price Prediction Dataset.csv')
```

```python
car_dataset.head()
```

Checking Number of Rows and Columns

```python
car_dataset.shape
```

```python
car_dataset.info()
```

Checking For Missing Values

```python
car_dataset.isnull().sum()
```

Checking the Distribution

```python
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encoding the Categorial Data

```python
car_dataset.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)
car_dataset.replace({'Seller_Type':{'Dealer':0, 'Individual':1}}, inplace=True)
car_dataset.replace({'Transmission':{'Manual':0, 'Automatic':1}}, inplace=True)
```

```python
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Splitting Data and Target

```python
X = car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y = car_dataset['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

Splitting Training and Testing Data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size = 0.1 , random_state=2)
```

Model Training

Linear Regression Model

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

Model Evaluation

```python
training_Data_Prediction = lin_reg_model.predict(X_train)
```

R Squared Error

```python
error_score_1 = metrics.r2_score(Y_train , training_Data_Prediction)
print("R squared Error :", error_score_1)
```

Testing the Model

```python
testing_Data_Prediction = lin_reg_model.predict(X_test)
```

```python
error_score_2 = metrics.r2_score(Y_test , testing_Data_Prediction)
print("R squared Error :", error_score_2)
```

```python
plt.scatter(Y_test, testing_Data_Prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price Vs Predicted Price")
```

Lasso Regression

```python
lasso_reg_model = Lasso()
```

```python
lasso_reg_model.fit(X_train,Y_train)
```

```python
training_Data_Prediction = lasso_reg_model.predict(X_train)
```

```python
error_score_1 = metrics.r2_score(Y_train , training_Data_Prediction)
print("R squared Error :", error_score_1)
```

```python
testing_Data_Prediction = lasso_reg_model.predict(X_test)
```

```python
error_score_2 = metrics.r2_score(Y_test , testing_Data_Prediction)
print("R squared Error :", error_score_2)
```

```python
plt.scatter(Y_test, testing_Data_Prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price Vs Predicted Price")
```
